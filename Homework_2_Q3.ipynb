```python
import cv2
import matplotlib.pyplot as plt
from pykuwahara import kuwahara

# Load the image
image = cv2.imread('COMEAU.jpg')  # Change to your image path
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for proper display

# Apply Kuwahara filters
filt1 = kuwahara(image, method='mean', radius=10)
filt2 = kuwahara(image, method='gaussian', radius=10)  # Default sigma is computed by OpenCV

# Save the filtered images
cv2.imwrite('lena-kfilt-mean.jpg', cv2.cvtColor(filt1, cv2.COLOR_RGB2BGR))  # Convert back to BGR for saving
cv2.imwrite('lena-kfilt-gaus.jpg', cv2.cvtColor(filt2, cv2.COLOR_RGB2BGR))  # Convert back to BGR for saving

# Display the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Kuwahara Mean Filter')
plt.imshow(filt1)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Kuwahara Gaussian Filter')
plt.imshow(filt2)
plt.axis('off')

plt.tight_layout()
plt.show()
```
